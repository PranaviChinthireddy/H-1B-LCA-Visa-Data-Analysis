```python
#Chavi
# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Identify missing values
missing_percent = df.isnull().mean() * 100

# Drop columns with more than 50% missing values
threshold = 50
cols_to_drop = missing_percent[missing_percent > threshold].index
df.drop(columns=cols_to_drop, inplace=True)

# Separate numeric and categorical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns
object_cols = df.select_dtypes(include=["object"]).columns

# Fill missing values intelligently
fill_values = {}

# For numeric columns, fill with median
for col in numeric_cols:
    fill_values[col] = df[col].median()

# For categorical columns, fill with mode (most frequent value)
for col in object_cols:
    if not df[col].mode().empty:
        fill_values[col] = df[col].mode().iloc[0]
    else:
        fill_values[col] = "No Data"  # Just in case mode calculation fails

# Apply missing value imputation
df.fillna(fill_values, inplace=True)

# Ensure no missing values remain
assert df.isnull().sum().sum() == 0, "There are still missing values!"

# Print final shape and sample output
print("Final shape after cleaning:", df.shape)
print(df.head(10))
```

```python
# Step 1: Filter for CERTIFIED H-1B cases and remove missing salary or cost data
approved_df = merged_df[merged_df['case_status'] == 'CERTIFIED']

# Step 2: Check missing values in salary and cost of living columns
print(approved_df[['wage_rate_of_pay_from', 'cost_of_living_2020']].isnull().sum())

# Step 3: Drop rows with missing values in the salary and cost of living columns
approved_df = approved_df.dropna(subset=['wage_rate_of_pay_from', 'cost_of_living_2020'])

# Step 4: Group by ZIP code (or city/state) to summarize
grouped_df = approved_df.groupby('worksite_postal_code').agg({
    'wage_rate_of_pay_from': 'mean',
    'cost_of_living_2020': 'mean',
    'case_number': 'count'  # counts number of approved petitions
}).rename(columns={'wage_rate_of_pay_from': 'avg_salary', 'case_number': 'num_approved'})

# Step 5: Add back city/state info (to show in results)
location_info = approved_df[['worksite_postal_code', 'worksite_city', 'worksite_state']].drop_duplicates()
grouped_df = grouped_df.reset_index().merge(location_info, on='worksite_postal_code', how='left')

# Step 6: Scale features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(grouped_df[['avg_salary', 'cost_of_living_2020', 'num_approved']])
grouped_df[['scaled_salary', 'scaled_col', 'scaled_approval']] = scaled

# Step 7: Score (you can tweak weights based on your focus)
grouped_df['score'] = (
    0.4 * grouped_df['scaled_salary'] -  # Higher salary is better
    0.3 * grouped_df['scaled_col'] +     # Lower cost of living is better
    0.3 * grouped_df['scaled_approval']  # More approvals is better
)

# Step 8: Sort by score to find the best cities
sorted_df = grouped_df.sort_values(by='score', ascending=False)

# Display the top 10 cities
print(sorted_df[['worksite_city', 'worksite_state', 'avg_salary', 'cost_of_living_2020', 'num_approved', 'score']].head(10))
```

```python
# Final Merge Pipeline: Safe column cleanup + random 20K sample

import pandas as pd
import numpy as np

# File paths
main_file = '/content/drive/MyDrive/Combined_LCA_Disclosure_Data_FY2020_to_FY2024.csv'
cost_file = '/content/drive/MyDrive/cost_of_living_by_zip_code (1).xlsx'
crime_file = '/content/drive/MyDrive/crimedata.csv'

# Step 1: Load all columns safely and clean
main_df = pd.read_csv(main_file, low_memory=False)
main_df.columns = main_df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Filter for Certified and Denied, and keep required columns
main_df = main_df[main_df['case_status'].isin(['Certified', 'Denied'])]
main_df = main_df[['case_number', 'job_title', 'worksite_city', 'worksite_state', 'worksite_postal_code', 'case_status']]
main_df.dropna(inplace=True)

# Standardize location fields
main_df['worksite_postal_code'] = main_df['worksite_postal_code'].astype(str).str.zfill(5)
main_df['worksite_city'] = main_df['worksite_city'].str.upper().str.strip()
main_df['worksite_state'] = main_df['worksite_state'].str.upper().str.strip()

# Step 3: Random sample of 20,000 rows (natural class distribution)
random_sample_df = main_df.sample(n=20000, random_state=42).reset_index(drop=True)

# Step 4: Load and clean cost of living dataset
cost_df = pd.read_excel(cost_file)
cost_df.columns = cost_df.columns.str.strip().str.lower().str.replace(" ", "_")
cost_df.rename(columns={'zip': 'worksite_postal_code'}, inplace=True)
cost_df['worksite_postal_code'] = cost_df['worksite_postal_code'].astype(str).str.zfill(5)

# Step 5: Load and clean crime dataset
crime_df = pd.read_csv(crime_file)
crime_df.columns = crime_df.columns.str.strip().str.lower().str.replace(" ", "_")
crime_df['community_name'] = crime_df['communityname'].str.upper().str.strip()
crime_df['state'] = crime_df['state'].str.upper().str.strip()
crime_df['community_name'] = crime_df['community_name'].str.replace('TOWNSHIP', '', regex=False)
crime_df['community_name'] = crime_df['community_name'].str.replace('CITY', '', regex=False)

# Step 6: Merge datasets
merged_df = pd.merge(random_sample_df, cost_df, on='worksite_postal_code', how='left')
merged_df = pd.merge(merged_df, crime_df,
                     left_on=['worksite_city', 'worksite_state'],
                     right_on=['community_name', 'state'], how='left')

# Step 7: Handle missing violent crime data
merged_df['missing_crime_data'] = merged_df['violentcrimesperpop'].isnull().astype(int)
merged_df['violentcrimesperpop'] = merged_df['violentcrimesperpop'].fillna(
    merged_df.groupby('worksite_state')['violentcrimesperpop'].transform('median')
)
merged_df['violentcrimesperpop'] = merged_df['violentcrimesperpop'].fillna(
    merged_df['violentcrimesperpop'].median()
)

# Step 8: Final cleaned dataset
final_df = merged_df[['case_number', 'job_title', 'worksite_city', 'worksite_state',
                      'worksite_postal_code', 'violentcrimesperpop', 'cost_of_living_2020',
                      'missing_crime_data', 'case_status']].reset_index(drop=True)

print("✅ Final dataset shape:", final_df.shape)
print(final_df['case_status'].value_counts())
print(final_df.head())

# Step 9: Export for download
output_path = "/content/final_merged_random_sample_20k.csv"
final_df.to_csv(output_path, index=False)

# For Google Colab: trigger download
from google.colab import files
files.download(output_path)
```

```python
# How can we predict whether a job is located in a high cost-of-living area based on socio-economic and job-related features?

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Prepare dataset
df = final_df.copy()
df = df[df['cost_of_living_2020'].notna()]  # Ensure no missing targets

#  New target: High vs. Low cost of living
threshold = df['cost_of_living_2020'].median()
df['high_cost_area'] = df['cost_of_living_2020'].apply(lambda x: 1 if x >= threshold else 0)

# Frequency encode high-cardinality features
for col in ['job_title', 'worksite_city', 'worksite_state', 'worksite_postal_code']:
    freq_map = df[col].value_counts().to_dict()
    df[col + '_freq'] = df[col].map(freq_map).fillna(1)

# Define features
features = ['violentcrimesperpop', 'missing_crime_data',
            'job_title_freq', 'worksite_city_freq', 'worksite_state_freq', 'worksite_postal_code_freq']

X = df[features].fillna(df[features].median())
y = df['high_cost_area']

# Step 2: Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Step 3: Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Apply SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Step 5: Train model
model = RandomForestClassifier(n_estimators=300, max_depth=10, random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# Step 6: Evaluate
y_probs = model.predict_proba(X_test_scaled)[:, 1]
y_pred = (y_probs >= 0.5).astype(int)

print(" Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print(" ROC AUC Score:", round(roc_auc_score(y_test, y_probs), 4))
print(" Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(" Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix Plot
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='crest',
            xticklabels=['Low Cost', 'High Cost'], yticklabels=['Low Cost', 'High Cost'])
plt.title("Cost of Living Classification (Random Forest)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
# Can we predict whether a job is located in a high-crime area based on work location and job-related attributes?

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Prep data
df = final_df.copy()
df = df[df['violentcrimesperpop'].notna()]

# New target: high-crime flag based on 75th percentile
crime_threshold = df['violentcrimesperpop'].quantile(0.75)
df['high_crime_area'] = df['violentcrimesperpop'].apply(lambda x: 1 if x >= crime_threshold else 0)

# Frequency encoding
for col in ['job_title', 'worksite_city', 'worksite_state', 'worksite_postal_code']:
    freq_map = df[col].value_counts().to_dict()
    df[col + '_freq'] = df[col].map(freq_map).fillna(1)

# Features
features = ['cost_of_living_2020', 'missing_crime_data',
            'job_title_freq', 'worksite_city_freq', 'worksite_state_freq', 'worksite_postal_code_freq']

X = df[features].fillna(df[features].median())
y = df['high_crime_area']

# Step 2: Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Step 3: Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Step 5: Train model
model = GradientBoostingClassifier(n_estimators=300, learning_rate=0.1, max_depth=5, random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# Step 6: Evaluate
y_probs = model.predict_proba(X_test_scaled)[:, 1]
y_pred = (y_probs >= 0.5).astype(int)

print(" Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print(" ROC AUC Score:", round(roc_auc_score(y_test, y_probs), 4))
print(" Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(" Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix Plot
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='flare',
            xticklabels=['Low Crime', 'High Crime'], yticklabels=['Low Crime', 'High Crime'])
plt.title("High-Crime Area Classification (Gradient Boosting)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

```python
# Q: Which features had the greatest impact on H-1B approval predictions ?

from xgboost import plot_importance, XGBClassifier
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Drop 'missing_crime_data' from the feature set if present
X_filtered = X.drop(columns=['missing_crime_data'], errors='ignore')

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, stratify=y, test_size=0.2, random_state=42)

# Save column names before scaling
feature_names = X_filtered.columns

# Scale and preserve column names
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=feature_names)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=feature_names)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Train XGBoost model
xgb = XGBClassifier(n_estimators=300, learning_rate=0.1, random_state=42)
xgb.fit(X_train_resampled, y_train_resampled)

# Plot top features (now without missing_crime_data)
plt.figure(figsize=(12, 8))
plot_importance(xgb, importance_type='gain', max_num_features=10, show_values=True,
                height=0.6, grid=False)
plt.title('Top Feature Importances for H-1B Approval Prediction ', fontsize=14, weight='bold')
plt.xlabel('Gain Score', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.tight_layout()
plt.show()
```
